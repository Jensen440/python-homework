# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:

* `whale_returns.csv`: Contains returns of some famous "whale" investors' portfolios.

* `algo_returns.csv`: Contains returns from the in-house trading algorithms from Harold's company.

* `sp500_history.csv`: Contains historical closing prices of the S&P 500 Index.

## Whale Returns

Read the Whale Portfolio daily returns and clean the data

```python
# Reading whale returns

whale_path = Path('Resources/whale_returns.csv')

whale_returns_csv = pd.read_csv(whale_path)
whale_returns_csv.set_index(pd.to_datetime(whale_returns_csv['Date'], infer_datetime_format=True), inplace=True)


# sorting data by date

whale_returns_csv.sort_index
```

```python
# Count nulls

whale_returns_csv.isnull().sum()
```

```python
# Drop nulls (and replace)

whale_returns_csv.dropna(inplace=True)
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# Reading algorithmic returns

algo_path = Path('Resources/algo_returns.csv')

algo_returns_csv = pd.read_csv(algo_path)

algo_returns_csv.set_index(pd.to_datetime(algo_returns_csv['Date'], infer_datetime_format=True), inplace=True)

algo_returns_csv.drop(columns=['Date'], inplace=True)

# sorting data by date

algo_returns_csv.sort_index
```

```python
# Count nulls

algo_returns_csv.isnull().sum()
```

```python
# Drop nulls (and replace)

algo_returns_csv.dropna(inplace=True)
```

## S&P 500 Returns

Read the S&P 500 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P 500 Closing Prices

sp500_path = Path('Resources/sp500_history.csv')


#sp500_csv = pd.read_csv(sp500_path)

#sp500_csv.set_index(pd.to_datetime(sp500_csv['Date'], infer_datetime_format=True), inplace=True)
#sp500_csv = pd.read_csv(sp500_path)
sp500_csv = pd.read_csv(sp500_path, index_col="Date", parse_dates=True, infer_datetime_format=True)

#sp500_csv['Date'] = pd.to_datetime(sp500_csv['Date'], infer_datetime_format=True, dayfirst = True)


# sorting data by date

sp500_csv.sort_index(ascending=True, inplace=True)

sp500_csv
```

```python
# Check Data Types

sp500_csv.dtypes

```

```python
# Fix Data Types (remove '$')

sp500_csv['Close'] = sp500_csv['Close'].str.replace('$', '').astype("float")
sp500_csv.sort_index(inplace=True)
sp500_csv
```

```python
# Calculate Daily Returns

sp500_daily_returns = sp500_csv.pct_change()
sp500_daily_returns.head()
```

```python
# Drop nulls

sp500_daily_returns.dropna(inplace=True)
sp500_daily_returns
```

```python
# Rename `Close` Column to be specific to this portfolio.

sp500_daily_returns.rename(columns = {"Close":"SP500"}, inplace=True)

sp500_daily_returns
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# Join Whale Returns, Algorithmic Returns, and the S&P 500 Returns into a single DataFrame with columns for each portfolio's returns.

combined_daily_returns = pd.concat([whale_returns_csv, algo_returns_csv, sp500_daily_returns], axis = "columns", join = "inner")

combined_daily_returns.drop(columns=['Date'], inplace=True)

combined_daily_returns.head()
```

In this section, you will calculate and visualize performance and risk metrics for the portfolios.

## Performance Anlysis

#### Calculate and Plot the daily returns.

```python
# Plot daily returns of all portfolios

combined_daily_returns.plot(title="Daily Returns",figsize = (20, 10))
```

## Risk Analysis

Determine the _risk_ of each portfolio:

1. Create a box plot for each portfolio. 
2. Calculate the standard deviation for all portfolios
4. Determine which portfolios are riskier than the S&P 500
5. Calculate the Annualized Standard Deviation

### Create a box plot for each portfolio


```python
# Box plot to visually show risk

combined_daily_returns.plot.box(title= "Risk", figsize=(20,10))
```

### Calculate Standard Deviations

```python
# Calculate the daily standard deviations of all portfolios

daily_std = combined_daily_returns.std()
daily_std
```

### Determine which portfolios are riskier than the S&P 500

```python
# Calculate  the daily standard deviation of S&P 500 (cannot use sp500_history.csv without reducing range of values)

sp500_std = daily_std["SP500"]
sp500_std

# Determine which portfolios are riskier than the S&P 500

risk_vs_sp500 =[daily_std > sp500_std]
risk_vs_sp500
```

### Calculate the Annualized Standard Deviation

```python
# Calculate the annualized standard deviation (252 trading days)
annual_std = combined_daily_returns.std() * np.sqrt(252)
annual_std.sort_values(inplace=True)
annual_std
```

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Calculate and plot the rolling standard deviation for all portfolios using a 21-day window
2. Calculate the correlation between each stock to determine which portfolios may mimick the S&P 500
3. Choose one portfolio, then calculate and plot the 60-day rolling beta between it and the S&P 500

### Calculate and plot rolling `std` for all portfolios with 21-day window

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window

combined_21day_std = combined_daily_returns.rolling(window=21).std()

# Plot the rolling standard deviation

combined_21day_std.plot(title="21 Day Rolling STD", figsize=(20,10))
```

### Calculate and plot the correlation

```python
# Calculate the correlation

combined_correlation = combined_daily_returns.corr()
combined_correlation

# Display de correlation matrix

sns.heatmap(combined_correlation, vmin=-1, vmax=1,annot=True)
```

### Calculate and Plot Beta for a chosen portfolio and the S&P 500

```python
# Calculate covariance of a single portfolio

algo1_covariance = combined_daily_returns['Algo 1'].rolling(window=30).cov(combined_daily_returns['SP500'])

# Calculate variance of S&P 500

sp500_variance = combined_daily_returns['SP500'].rolling(window=30).var()

# Computing beta

algo1_beta = algo1_covariance / sp500_variance

# Plot beta trend

algo1_beta_graph = algo1_beta.plot(title="Algo 1 30-Day Rolling Beta", figsize=(20,10))
```

## Rolling Statistics Challenge: Exponentially Weighted Average 

An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the [`ewm`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.ewm.html) with a 21-day half life for each portfolio, using standard deviation (`std`) as the metric of interest.

```python
# Use `ewm` to calculate the rolling window

rolling_ewm = combined_daily_returns.ewm(halflife=21, adjust=True).std().plot(title="Exponentially Weighted Average", figsize=(20, 10))

```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios

sharpe_ratios = (combined_daily_returns.mean()*252/(combined_daily_returns.std()*np.sqrt(252)))
sharpe_ratios
```

```python
# Visualize the sharpe ratios as a bar plot

sharpe_ratios.plot(kind="bar", title="Sharpe Ratios", figsize=(20,10))
```

## Choose 3-5 custom stocks with at last 1 year's worth of historic prices and create a DataFrame of the closing prices and dates for each stock.

```python
# Reading data from 1st stock

gme_data = Path("Resources/GME_data.csv")
gme_df = pd.read_csv(gme_data, index_col="Date", parse_dates=True, infer_datetime_format=True)
gme_df.head()
```

```python
# Reading data from 2nd stock

msft_data = Path("Resources/MSFT_data.csv")
msft_df = pd.read_csv(msft_data, index_col="Date", parse_dates=True, infer_datetime_format=True)
msft_df.head()
```

```python
# Reading data from 3rd stock

tsla_data = Path("Resources/TSLA_data.csv")
tsla_df = pd.read_csv(tsla_data, index_col="Date", parse_dates=True, infer_datetime_format=True)
tsla_df.head()
```

```python
# Combine all stocks in a single DataFrame

combined_portfolio_df = pd.concat([gme_df, msft_df, tsla_df], axis="columns", join="inner")
```

```python
# Reset Date index

combined_portfolio_df.sort_index(inplace=True)
```

```python
# Reorganize portfolio data by having a column per symbol

combined_portfolio_df.columns = ["GME", "MSFT", "TSLA"]

combined_portfolio_df.head()
```

```python
# Calculate daily returns

portfolio_daily_returns = combined_portfolio_df.pct_change()

# Drop NAs

portfolio_daily_returns.dropna(inplace=True)

# Display sample data

portfolio_daily_returns.head()
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Set weights

weights = [1/3, 1/3, 1/3]

# Calculate portfolio return

portfolio_weighted_returns = portfolio_daily_returns.dot(weights)

# Display sample data

portfolio_weighted_returns.sort_index(inplace=True)

# Drop time from Date/Time

portfolio_weighted_returns.index = portfolio_weighted_returns.index.normalize()

portfolio_weighted_returns.head()
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
# Join your returns DataFrame to the original returns DataFrame

combined_portfolio_2 = pd.concat((portfolio_weighted_returns, combined_daily_returns), axis = "columns", join = "inner")

combined_portfolio_2.rename(columns = {0:"Custom Portfolio"}, inplace = True)

combined_portfolio_2.head()
```

```python
# Only compare dates where return data exists for all the stocks (drop NaNs)

combined_portfolio_2.dropna(inplace=True)

combined_portfolio_2.head()
```

## Re-run the risk analysis with your portfolio to see how it compares to the others

### Calculate the Annualized Standard Deviation

```python
# Calculate the annualized `std`

portfolio_std_annual = combined_portfolio_2.std() * np.sqrt(252)
portfolio_std_annual.sort_values(inplace=True)
portfolio_std_annual
```

### Calculate and plot rolling `std` with 21-day window

```python
# Calculate rolling standard deviation

portfolio_rolling_std21 = combined_portfolio_2[["Custom Portfolio"]].rolling(window = 21).std()

# Plot rolling standard deviation
portfolio_rolling_std21.plot(figsize = (20,10), title = "Custom Portfolio 21 Day STD")

```

### Calculate and plot the correlation

```python
# Calculate and plot the correlation

combined_correlation2 = combined_portfolio_2.corr()

sns.heatmap(combined_correlation2, vmin=-1, vmax=1, annot=True)

```
